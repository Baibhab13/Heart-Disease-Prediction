# Importing the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data collection and processing

# Splitting the feature and target

```python
x = hd.drop(columns = 'target', axis = 1)
y = hd['target']
```

```python
print(x)
```

```python
print(y)
```

Splitting data into training and testing data


```python
x_t,x_ts,y_t,y_ts= train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print("Size of the x train data: ",x_t.shape)
print("Size of the x test data: ",x_ts.shape)
print("Size of the y train data: ",y_t.shape)
print("Size of the y test data: ",y_ts.shape)
```

# Model training

Logistic regression

```python
model = LogisticRegression()
```

```python
# Training the LogisticRegression model with Training data
model.fit(x_t,y_t)
```

# Model Evaluation

Accuracy score evaluation

```python
# accuracy on training data
x_t_p = model.predict(x_t)
training_data_accuracy = accuracy_score(x_t_p,y_t)
```

```python
print('Accuracy on training data:', training_data_accuracy)
```

```python
# accuracy on training data
x_ts_p = model.predict(x_ts)
testing_data_accuracy = accuracy_score(x_ts_p,y_ts)
```

```python
print('Accuracy on testing data:', testing_data_accuracy)
```

# Building a predictive system

```python
input_data = (62,0,0,138,294,1,1,106,0,1.9,1,3,2)

# change the input datato a numpy array
imput_data_as_np=np.array(input_data)

# reshape the numpy array as we are predicting for only one instance
input_data_reshaped = imput_data_as_np.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
